#Question 5: what is Time required per WI per user in L1 ?

```python
L1_TPWI = df.groupby(["Pass#1 Complete By(User name)"])['Duration for Pass#1 Complete (Seconds)'].sum()
L1_TPWI = pd.DataFrame(L1_TPWI)
L1_TPWI['Workitem ID'] = L1_Users_Workitems['Workitem ID']
L1_TPWI['Duration for Pass#1 Complete (Minuets)'] = L1_TPWI['Duration for Pass#1 Complete (Seconds)'] / 60
L1_TPWI = L1_TPWI.reset_index()
L1_TPWI['TimePerWI_Seconds'] = L1_TPWI['Duration for Pass#1 Complete (Seconds)']/L1_TPWI['Workitem ID']
#L1_TPF.drop('9')
L1_TPWI = L1_TPWI.drop([df.index[9]])
L1_TPWI = L1_TPWI.drop([df.index[15]])
L1_TPWI['TimePerWI_Minuets'] = L1_TPWI['TimePerWI_Seconds'] / 60
L1_TPWI.head(5)
```

```python
#df.plot(x="Pass#1 Complete By(User name)", y=["Duration for Pass#1 Complete (Seconds)", "Workitem ID"], kind="bar")
sns.set(style='white')
plt.figure(figsize=(15,6))
bar_plot = sns.barplot(x='Pass#1 Complete By(User name)', y="TimePerWI_Minuets", data = L1_TPWI)
plt.xlabel('Users')
plt.ylabel('Time Required/Work Item')
plt.xticks(rotation=90)
plt.show()
```

#Question 6: what is Time required per WI per user in L2 ?

```python
L2_TPWI = df.groupby(["Pass#2 Complete By(User name)"])['Duration for Pass#2 Complete (Seconds)'].sum()
L2_TPWI = pd.DataFrame(L2_TPWI)
L2_TPWI['Workitem ID'] = L2_Users_Workitems['Workitem ID']
L2_TPWI['Duration for Pass#1 Complete (Minuets)'] = L2_TPWI['Duration for Pass#2 Complete (Seconds)'] / 60
L2_TPWI = L2_TPWI.reset_index()
L2_TPWI['TimePerWI_Seconds'] = L2_TPWI['Duration for Pass#2 Complete (Seconds)']/L2_TPWI['Workitem ID']
L2_TPWI['TimePerWI_Minuets'] = L2_TPWI['TimePerWI_Seconds'] / 60
L2_TPWI.head(5)
```

```python
sns.set(style='white')
plt.figure(figsize=(15,6))
bar_plot = sns.barplot(x='Pass#2 Complete By(User name)', y="TimePerWI_Minuets", data = L2_TPWI)
plt.xlabel('Users')
plt.ylabel('Time Required/Work Item')
plt.xticks(rotation=90)
plt.show()
```

#Work Item level analysis does not provide an accurate work efficiency of users since one workitem can have uncertain amount of datafileds to label which is dpends upon the document type. To get precise work efficiency of users lets perform field level analysis for L1 and L2 users.

```python
L1_Field =  df[["Pass#1 Complete By(User name)", "Pass#1 Fields Validated By User", "Duration for Pass#1 Complete (Seconds)"]]
L1_Field = L1_Field.groupby(["Pass#1 Complete By(User name)"]).sum()
L1_Field = L1_Field.reset_index()
L1_Field.head(5)
```

```python
#lets find out total time taken by user to validaate per field in seconds
L1_Field_TPF = L1_Field
L1_Field_TPF['TimePerField (seconds)'] = L1_Field_TPF['Duration for Pass#1 Complete (Seconds)']/L1_Field_TPF['Pass#1 Fields Validated By User']
L1_Field_TPF.head(5)
```

```python
L1_Field_TPF.plot(x="Pass#1 Complete By(User name)", y= "TimePerField (seconds)", kind="bar",figsize=(15,6))
plt.xlabel('Users')
plt.ylabel('Time Per Field')
plt.show()
```
